```python
import geopandas as gpd
import pandas as pd
# from geographiclib.geodesic import Geodesic
# from shapely.geometry import MultiLineString
# from shapely.geometry import Point
import numpy as np

import folium
```

```python
country_boundaries = gpd.read_file('./data/world-administrative-boundaries.geojson')
```

```python
my_travel_df = pd.read_excel('./data/Travel History.xlsx')
my_travel_df.head()
```

```python
iso3_unique = my_travel_df.iso3.unique()
```

```python
country_visited = country_boundaries[country_boundaries.iso3.isin(iso3_unique)]
```

```python
m = folium.Map([40, 50], zoom_start=2, tiles="cartodbpositron")

for country_code in iso3_unique:
    country_gdf = country_boundaries[country_boundaries.iso3 == country_code]
    df_show = my_travel_df[my_travel_df.iso3 == country_code].drop(columns=['iso3'])
    df_show.index = np.arange(len(df_show))+1
    html = df_show.to_html(
        classes="table table-striped table-hover table-condensed table-responsive"
    )
    folium.GeoJson(
        country_gdf,
        # name="lines",
        style_function=lambda x: {
            "fillColor": "orange",
            "color": 'black',
            "weight": 1.0,
            "opacity": 0.8,
            "fillOpacity": 0.5,
        },
        popup=folium.Popup(html=html,max_width="300"),
        highlight_function=lambda x: {"fillOpacity": 0.9},
        # zoom_on_click=True,
        ).add_to(m)
m
```
